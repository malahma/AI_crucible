```python
#17 What is the result of the following expression?

print(0 * np.nan)             # Result: nan
print(np.nan == np.nan)       # Result: False
print(np.inf > np.nan)        # Result: False
print(np.nan - np.nan)        # Result: nan
print(np.nan in set([np.nan]))# Result: True
print(0.3 == 3 * 0.1)         # Result: False
```

```python
#19 Create an 8x8 matrix and fill it with a checkerboard pattern
Z = np.zeros((8,8), dtype=int)
Z[1::2, ::2] = 1
Z[::2, 1::2] = 1
print(Z)
```

```python
#20 Consider a (6,7,8) shape array, what is the index (x,y,z) of the 100th element

res = np.unravel_index(100, (6,7,8))
result = tuple(int(i) for i in res)
print(result)
```

```python
#22 Normalize a 5x5 random matrix
Z = np.random.random((5,5))
Z = (Z - Z.min()) / (Z.max() - Z.min())
Z_int = (Z * 10).astype(int)

print(Z_int)
```

```python
# 23. Create a custom dtype that describes a color as four unsigned bytes (RGBA)
color_layout = [
    ("r", np.uint8),
    ("g", np.uint8), 
    ("b", np.uint8), 
    ("a", np.uint8) 
]

color = np.dtype(color_layout)

print(color)
```

```python
 #24 Multiply a 5x3 matrix by a 3x2 matrix (real matrix product)
import numpy as np
A = np.ones((5, 3))
B = np.ones((3, 2))
Z = (A @ B).astype(int)

print(Z)
```

```python
#25 Given a 1D array, negate all elements which are between 3 and 8, in place.
Z = np.arange(11)
Z[(3 < Z) & (Z <= 8)] *= -1

print(Z)
```

```python
# 27. Consider an integer vector Z, which of these expressions are legal?
import numpy as np
Z = np.arange(3)

print(Z**Z)       
print(2 << Z >> 2) 
print(Z <- Z)    
print(1j*Z)      
print(Z/1/1)      
```
