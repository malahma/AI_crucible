```python
#35 35. How to compute ((A+B)*(-A/2)) in place (without copy)?
A = np.ones(3) * 1
B = np.ones(3) * 2

np.add(A, B, out=B)

np.divide(A, -2, out=A)

np.multiply(A, B, out=A)

print(A)
```

```python
#44 Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates
import numpy as np

Z = np.random.random((10, 2))

X, Y = Z[:, 0], Z[:, 1]

R = np.sqrt(X**2 + Y**2)

T = np.arctan2(Y, X)

print("Polar Coordinates (R, Theta):")
print(np.column_stack((R, T)))
```
